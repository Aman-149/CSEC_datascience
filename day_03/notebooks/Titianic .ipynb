```python
import pandas as pd

# Load Titanic dataset
titanic = pd.read_csv("/kaggle/input/datasets/amanuelfikadu/titanic-dataset/Titanic-Dataset.csv")
```

Exploration

```python
print(titanic.head())
print(titanic.info())
print(titanic.describe())
```

# Data Cleaning****

```python
# Fill missing values
titanic["Age"] = titanic["Age"].fillna(titanic["Age"].median())
titanic["Embarked"] = titanic["Embarked"].fillna(titanic["Embarked"].mode()[0])

# Remove duplicates
titanic.drop_duplicates(inplace=True)

# Final check
print(titanic.isnull().sum())
```

## Survival rate by gender

```python
survival_gender = titanic.groupby("Sex")["Survived"].mean()
print(survival_gender)
```

## Survival rate by class

```python
survival_class = titanic.groupby("Pclass")["Survived"].mean()
print(survival_class)
```

## Average age per class

```python
avg_age_class = titanic.groupby("Pclass")["Age"].mean()
print(avg_age_class)
```

## Survival rate by age group

```python
# Create age groups
bins = [0, 12, 18, 60, 100]
labels = ["Child", "Teen", "Adult", "Senior"]

titanic["AgeGroup"] = pd.cut(titanic["Age"], bins=bins, labels=labels)

survival_age_group = titanic.groupby("AgeGroup", observed=False)["Survived"].mean()
print(survival_age_group)
```

# Filtering

## Female passengers who survived

```python
female_survived = titanic[(titanic["Sex"] == "female") & (titanic["Survived"] == 1)]
print(female_survived.head())
```

## Children who survived

```python
children_survived = titanic[(titanic["Age"] < 12) & (titanic["Survived"] == 1)]
print(children_survived.head())
```

## 1st class passengers with high survival

```python
first_class_survived = titanic[(titanic["Pclass"] == 1) & (titanic["Survived"] == 1)]
print(first_class_survived.head())
```

# Insights 
1. Who was more likely to survive?                                                                  
Females had a much higher survival rate than males.
2. Did class affect survival?
Yes
1st class → Highest survival
2nd class → Medium
3rd class → Lowest
3. Were children prioritized?
Yes.
Children had higher survival rates than adults, showing a "women and children first" policy.
4. What combination had the highest survival rate?
Female + 1st Class + Young age (Child/Teen)
